```python
import pandas as pd

df = pd.read_csv('simpsons_paradox_covid.csv')
```

```python
df.head()
```

```python
#Gesamtzahl Fälle und Todesfälle nach Impfstatus
summary = df.groupby("vaccine_status")["outcome"].value_counts().unstack().fillna(0)
#Sterberate = Todesfälle / Gesamtfälle
summary["Sterberate"] = summary["death"] / (summary["death"] + summary["survived"])
print(summary)
```

```python
import altair as alt
#Sterberaten berechnen
summary = (
    df
    .groupby("vaccine_status")["outcome"]
    .value_counts()
    .unstack(fill_value=0)
    .reset_index()
)
summary["death_rate"] = summary["death"] / (summary["death"] + summary["survived"])

#Altair-Grafik erstellen: Sterberate nach Impfstatus
chart = (
    alt.Chart(summary)
    .mark_bar(cornerRadiusTopLeft=3, cornerRadiusTopRight=3)
    .encode(
        x=alt.X("vaccine_status:N", title="Impfstatus"),
        y=alt.Y("death_rate:Q", title="Sterberate", axis=alt.Axis(format="%")),
        color=alt.Color("vaccine_status:N", legend=None)
    )
    .properties(
        width=200,
        height=300,
        title="Sterberate nach Impfstatus"
    )
)

chart
```

```python
#Sterberaten berechnen pro Altersgruppe und Impfstatus
summary = (
    df
    .groupby(["age_group", "vaccine_status"])["outcome"]
    .value_counts()
    .unstack(fill_value=0)
    .reset_index()
)

#Berechne death_rate = Todesfälle / (Todesfälle + Überlebende)
summary["death_rate"] = summary["death"] / (summary["death"] + summary["survived"])

#Altair-Chart: für jede Altersgruppe nebeneinander die Balken für geimpft/ungeimpft
chart = (
    alt.Chart(summary)
    .mark_bar(cornerRadiusTopLeft=3, cornerRadiusTopRight=3)
    .encode(
        column=alt.Column("age_group:N", title="Altersgruppe"),
        x=alt.X("vaccine_status:N", title="Impfstatus"),
        y=alt.Y("death_rate:Q", title="Sterberate", axis=alt.Axis(format="%")),
        color=alt.Color("vaccine_status:N", legend=None)
    )
    .properties(
        width=200,
        height=300,
        title="Sterberate nach Altersgruppe und Impfstatus"
    )
)

chart
```
