```python
import pandas as pd
df = pd.read_csv('car_repairs.csv')
df.head()
```

```python

import altair as alt


# Reparatur-Anzahl pro Modell berechnen
repair_counts = (
    df[df['Repaired'] == True]
    .groupby('Model')
    .size()
    .reset_index(name='repairs')
)

# Balkendiagramm mit Altair
chart = (
    alt.Chart(repair_counts)
    .mark_bar()
    .encode(
        x=alt.X('Model:N', title='Modell'),
        y=alt.Y('repairs:Q', title='Anzahl Reparaturen'),
        color=alt.Color('Model:N', legend=None)
    )
    .properties(
        title='Absolute Anzahl Reparaturen pro Modell'
    )
)

chart.show()
```

```python
# Modell-Anzahlen berechnen
model_counts = (
    df
    .groupby('Model')
    .size()
    .reset_index(name='count')
)

# Balkendiagramm auf Basis des aggregierten DataFrames
chart = (
    alt.Chart(model_counts)
    .mark_bar()
    .encode(
        x=alt.X('Model:N', title='Modell'),
        y=alt.Y('count:Q', title='Anzahl Fahrzeuge'),
        color=alt.Color('Model:N', legend=None)
    )
    .properties(title='Verteilung der Fahrzeuge nach Modell')
)

chart.show()
```

```python
# Reparatur-Rate pro Modell vorab berechnen
repair_rates = (
    df
    .groupby('Model')['Repaired']
    .mean()
    .reset_index(name='repair_rate')
)

# Balkendiagramm auf Basis des kleinen DataFrames
chart = (
    alt.Chart(repair_rates)
    .mark_bar()
    .encode(
        x=alt.X('Model:N', title='Modell'),
        y=alt.Y('repair_rate:Q', title='Reparatur-Rate', axis=alt.Axis(format='%')),
        color=alt.Color('Model:N', legend=None)
    )
    .properties(title='Reparatur-Rate pro Modell')
)

chart.show()
```

```python
rate_chart = (
    alt.Chart(repair_rates)
    .mark_bar(size=60)
    .encode(
        x=alt.X('Model:N', title='Modell'),
        y=alt.Y('repair_rate:Q', title='Reparatur-Rate', axis=alt.Axis(format='%')),
        color=alt.Color('Model:N', legend=None)
    )
    .properties(width=400, title='Reparatur-Rate pro Modell')  # Breiter machen
)

rate_chart.show()
```

```python
# CSV einlesen
df = pd.read_csv('car_repairs.csv')

# Reparatur-Anzahl pro Modell berechnen
repair_counts = (
    df[df['Repaired'] == True]
    .groupby('Model')
    .size()
    .reset_index(name='repairs')
)

# Balkendiagramm mit Altair – im Stil von rate_chart
chart = (
    alt.Chart(repair_counts)
    .mark_bar(size=60)  # breitere Balken
    .encode(
        x=alt.X('Model:N', title='Modell'),
        y=alt.Y('repairs:Q', title='Anzahl Reparaturen'),
        color=alt.Color('Model:N', legend=None)
    )
    .properties(
        width=400,
        title='Absolute Anzahl Reparaturen pro Modell'
    )
)

chart.show()
```
